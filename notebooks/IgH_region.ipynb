```python
import pandas as pd
```

```python
combined = pd.read_excel("combined_notExtend.xlsx" , sheet_name=None)
sheets = list(combined.keys())
combined_loci = pd.DataFrame()

for name in sheets:
    for chrom in combined[name]['chrom'].unique():
        df = pd.DataFrame([[chrom, combined[name].loc[combined[name]['chrom']==chrom , 'start'].min(),
                          combined[name].loc[combined[name]['chrom']==chrom , 'end'].max()]], index=[name])
        combined_loci = pd.concat([combined_loci ,df])
combined_loci = combined_loci.reset_index()
combined_loci.columns = ['Species', 'chrom', 'start', 'end']
combined_loci
```

```python
species = pd.read_excel("species.xlsx")
species = dict(zip(species['common_name'], species['species']))
combined_loci= combined_loci.replace({"Species": species})
combined_loci.to_csv('IgH_loci.txt', sep='\t', index=False)
```
